```python
import pandas as pd
import yfinance as yf
import os

# Define the start and end dates
start_date = "2023-11-17"
end_date = "2023-11-30"

# Empty DataFrame to store data
data_df = pd.DataFrame()

# Define the folder paths
folder_path = 'C:/Users/almas/Downloads/archive17/Data/StockHistory'
destination_path = 'C:/Users/almas/Downloads/archive17/Data/API Extracted Data'

# Get a list of all CSV files in the destination folder
destination_files = os.listdir(destination_path)

# Iterate through each CSV file in the source folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        # Check if the file already exists in the destination folder
        if filename not in destination_files:
            # Construct API URL
            ticker_symbol = filename.split('.')[0]

            # Download data from Yahoo Finance
            data = yf.download(ticker_symbol, start=start_date, end=end_date)

            # Extract daily data and filter by date range
            daily_data = []
            for date, values in data.iterrows():
                if date >= pd.to_datetime(start_date) and date <= pd.to_datetime(end_date):
                    daily_data.append({
                        "date": date,
                        "open": values["Open"],
                        "high": values["High"],
                        "low": values["Low"],
                        "close": values["Close"],
                        "volume": values["Volume"],
                    })

            # Convert filtered data to DataFrame
            daily_df = pd.DataFrame(daily_data)

            # Append data for each ticker to the main DataFrame
            data_df = pd.concat([data_df, daily_df])

            # Save data to CSV file
            data_df.to_csv(os.path.join(destination_path, filename), index=False)

# Print final data
print("Data collected and saved to CSV files.")
```
